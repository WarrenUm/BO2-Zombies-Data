```python
#makin some plots
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
dow = pd.read_csv('data/dow_zombie_stats')
perry = pd.read_csv('data/perry_zombie_stats')
thornton = pd.read_csv('data/thornton_zombie_stats')
```

```python
dow.drop(columns='Unnamed: 0',inplace=True)
perry.drop(columns='Unnamed: 0',inplace=True)
thornton.drop(columns='Unnamed: 0',inplace=True)
```

```python
sns.pairplot(dow)
```

```python
sns.set_theme(context='notebook',
    style='darkgrid',
    palette='tab10',
    font='sans-serif',
    font_scale=2,
    rc=None,)
fig, ax = plt.subplots(figsize=(10,10))

sns.regplot(x = "Rounds", y = "headshots", data = dow, ax=ax,fit_reg=True,marker='*')
sns.regplot(x = "Rounds", y = "headshots", data = thornton, ax=ax,fit_reg=True,marker='^')
sns.regplot(x='Rounds',y="headshots",data=perry,ax=ax,fit_reg=True,marker='p')
ax.legend(['Dow','Thornton','Perry'])
ax.set_title("Headshots vs Number of Rounds")
ax.set_ybound(lower=0)
```

```python
sns.set_theme(context='notebook',
    style='darkgrid',
    palette='tab10',
    font='sans-serif',
    font_scale=2,
    rc=None,)

fig, ax = plt.subplots(ncols=3,figsize=(16,8),sharey=True)

sns.histplot(x='kills',data=dow,stat='count',ax=ax[0],bins=30,alpha=0.35,color='blue',element='bars',kde=True)
sns.histplot(x='kills',data=thornton,stat='count',ax=ax[1],bins=30,alpha=0.35,color='orange',element='bars',kde=True)
sns.histplot(x='kills',data=perry,stat='count',ax=ax[2],bins=30,alpha=0.35,color='green',element='bars',kde=True)
ax[0].set_xlabel('Dow Kills')
ax[1].set_xlabel('Thornton Kills')
ax[2].set_xlabel('Perry Kills')
ax[1].set_title('Counts of Kill Quantities')
```
